```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
home = pd.read_csv("https://bit.ly/3FC52AL")
```

```python
data = home[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE','DIS', 'RAD', 'TAX', 'PTRATIO', 'LSTAT']].to_numpy()
target = home['CAT_MEDV'].to_numpy()
```

```python
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.3, random_state= 45)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_jobs=-1, random_state=45)
```

```python
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.round((np.mean(scores['train_score']), np.mean(scores['test_score'])), 2))
```

```python
rf.fit(train_input, train_target)
print(np.round(rf.feature_importances_, 2))
```

# 12

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print(confusion_matrix(test_target, rf.predict(test_input)))
print(classification_report(test_target, rf.predict(test_input)))
```

```python
wine = pd.read_csv("http://bit.ly/wine_csv_data")
```

```python
wine
```

```python
white = (wine['class'] == 1)
```

```python
whitewine = wine[white]
```

```python
whitewine.describe()
```

```python
input = whitewine[['alcohol', 'sugar', 'pH']].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(input)
scaled = ss.transform(input)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 4, random_state = 47)
km.fit(scaled)
```

```python
print(np.unique(km.labels_, return_counts=True))
```

```python
print(km.transform(scaled[100:101]))
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
mart = pd.read_csv("https://bit.ly/3RDmQhN")
```

```python
itemsets = apriori(mart, min_support = 0.15, use_colnames = True)
```

```python
rules = association_rules(itemsets, metric='confidence', min_threshold= 0.6)
```

```python
results = rules.sort_values(by=['lift'], ascending = False)
```

```python
print(np.round(results, 2))
```
