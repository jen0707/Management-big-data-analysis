```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
housing_df = pd.read_csv('https://raw.githubusercontent.com/Sharon12Lee/test/main/BostonHousing.csv')
housing_df = housing_df.rename(columns={'CAT. MEDV' : 'CAT_MEDV'})
housing_df.head(9)
```

```python
new_housing_df = housing_df.drop(['CHAS', 'CAT_MEDV'], axis = 1)

new_housing_df.head(9)
```

```python
new_housing_df.describe()
```

```python
corr = new_housing_df.corr(method='pearson')
corr.round(2)
```

```python
from scipy.stats import spearmanr
CHAS_CAT_MEDV_correlation, p_value = spearmanr(housing_df['CHAS'], housing_df['CAT_MEDV'])

print("Spearman correlation between CHAS and CAT_MEDV: ", round(CHAS_CAT_MEDV_correlation, 2))
print("P-value: ", round(p_value, 3))
```

```python
import seaborn as sns
sns.heatmap(corr, xticklabels=corr.columns, yticklabels= corr.columns)
plt.title('Heatmap of correlation Table of Factors about BostonHousing', fontsize=12)
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels= corr.columns, vmin=-1, vmax=1, cmap='RdBu')
plt.title('Heatmap of correlation Table of Factors about BostonHousing', fontsize=12)
```
