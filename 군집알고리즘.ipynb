```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = np.load('fruits_300.npy')
```

```python
print(fruits.shape) # 넘파이 3차원 배열
```

```python
print(fruits[0, 0, : ]) # 첫 번째 이미지의 첫 번째 행의 모든 열 출력
```

```python
plt.imshow(fruits[0], cmap='gray') # 첫 번째 이미지 그리기
plt.show() # 0에 가까울 수로 검고 높은 값일 수록 밝다 (0~255)
```

# matplotlib의 imshow() 함수
- imshow()는 array의 값들을 색으로 환산해 이미지의 형태로 보여줌
- imshow()는 주어진 array에서 최대값을 1, 최소값을 0으로 대응해서 전체적인 값을 동일한 비율로 조절하여 0~1 사이의 숫자로 변환한 후에 이미지를 나타내 줌
- 나타나는 이미지 색상을 변경하려면
  - matplotlib의 cm에 있는 colormap 이용
  - imshow()의 cmap 매개변수(옵션)을 변경해주면 됨   
    (imshow()에서 사용하는 colormap의 default: viridis
    = 최소값이 보라, 최대값이 노랑으로 대응되는 colormap)

```python
plt.imshow(fruits[0], cmap='gray_r') # R (명암 반전)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2) # 하나의 행, 두 개의 열
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
plt.show()
```
